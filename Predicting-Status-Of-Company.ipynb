```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"] = (20,10)
```

```python
df1 = pd.read_csv("Q2_data_set - Q2_data_set.csv")
df1
```

```python
df1['COMPANY_STATUS'].unique()
```

```python
dff = []
def abc(df):
    for i in range(0,len(df)):
        if df[i]=='ACTIVE':
            dff.append(0)
        elif df[i]=='ACTIVE IN PROGRESS':
            dff.append(1)
        elif df[i]=='DORMANT':
            dff.append(2)
        elif df[i]=='STRIKE OFF':
            dff.append(3)
        elif df[i]=='AMALGAMATED':
            dff.append(4)
        elif df[i]=='CONVERTED TO LLP':
            dff.append(5)
        elif df[i]=='CONVERTED TO LLP AND DISSOLVED':
            dff.append(6)
```

```python
abc(df1['COMPANY_STATUS'])
```

```python
len(df1['COMPANY_STATUS'])
```

```python
len(dff)
```

```python
d = pd.DataFrame(dff)
```

```python
d[0]
```

```python
df1['COMPANY_STATUS'] = pd.concat([d[0]])
```

```python
df1
```

```python
df1['PAIDUP_CAPITAL'] = df1['PAIDUP_CAPITAL'].replace('-',np.nan, regex=True)
df1['AUTHORIZED_CAPITAL'] = df1['AUTHORIZED_CAPITAL'].replace('-',np.nan, regex=True)
```

```python
df2 = df1
df2
```

```python
df2.dropna(inplace=True)
```

```python
#Creating the Text into numeric(dummy values) so that machine can understand
principal_dummies = pd.get_dummies(df1['PRINCIPAL_BUSINESS_ACTIVITY'])
company_status_dummies = pd.get_dummies(df1['COMPANY_STATUS'])
company_category_dummies = pd.get_dummies(df1['COMPANY_CATEGORY'])
company_class_dummies = pd.get_dummies(df1['COMPANY_CLASS'])
```

```python
df2 = df1.drop(['CORPORATE_IDENTIFICATION_NUMBER','COMPANY_NAME','COMPANY_CLASS','DATE_OF_REGISTRATION','COMPANY_CATEGORY','REGISTERED_OFFICE_ADDRESS','REGISTRAR_OF_COMPANIES','SUB_CATEGORY','PRINCIPAL_BUSINESS_ACTIVITY'],axis = 'columns')
```

```python
#Converting into string because there are some float values and we can't create graph from mismatch elements
df2 = df2.astype(str)
df2
```

```python
#Now converting all to float for making models
df2 = df2.astype(float)
```

```python
df3 = pd.concat([principal_dummies,company_category_dummies,df2[['PAIDUP_CAPITAL','AUTHORIZED_CAPITAL','COMPANY_STATUS']]],axis='columns')
```

```python
df3 = df3.dropna()
df3
```

```python
X = df3.drop(['COMPANY_STATUS'],axis = 'columns')
X
```

```python
Y = df3['COMPANY_STATUS']
Y
```

```python
from sklearn import preprocessing
from sklearn import utils

lab_enc = preprocessing.LabelEncoder()
encoded = lab_enc.fit_transform(Y)
```

```python
encoded
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, encoded, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression()
logisticRegr.fit(x_train, y_train)
```

```python
logisticRegr.score(x_test,y_test)
```

```python
from sklearn.svm import SVC
from sklearn import preprocessing
model = SVC()
model = model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv1 = ShuffleSplit(n_splits = 5, test_size = 0.2, random_state = 0)

cross_val_score(model, X,Y, cv = cv1)
```

```python
y_predicted = model.predict(x_test)
```

```python
y_test,y_predicted
```

```python
plt.hist(y_test,color="red")
plt.hist(y_predicted,color="green")
```

```python
data = {'y_Actual':    y_test,
        'y_Predicted': y_predicted
        }

df = pd.DataFrame(data, columns=['y_Actual','y_Predicted'])
print (df)
```

```python
df_confusion = pd.crosstab(df['y_Actual'], df['y_Predicted'], rownames=['Actual'], colnames=['Predicted'])
df_confusion
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
sn.heatmap(df_confusion, annot=True)
plt.show()
```
